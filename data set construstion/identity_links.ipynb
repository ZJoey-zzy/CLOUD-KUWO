```python
import pandas as pd
import numpy as np
from pandasql import sqldf
from collections import Counter
```

```python
cloud = pd.read_csv("./prossed_cloud.csv", encoding='utf-8')
kuwo = pd.read_csv("./prossed_kuwo.csv", encoding='utf-8')
```

```python
cloud
```

```python
pysqldf = lambda query: sqldf(query, globals())
```

```python
q11 = """select distinct cloud.album_id as album_id1, kuwo.album_id as album_id2
from cloud,kuwo
where cloud.album_name = kuwo.album_name"""
```

```python
r11 = pysqldf(q11)
r11
```

```python
filter_artist =['伍佰 &amp; China Blue','七叔-叶泽浩','丫蛋蛋','丫蛋蛋/沈虫虫']
filter_list = []
for id1,id2 in r11.values.tolist():
    a1 =  cloud.loc[cloud.album_id == id1].iloc[0]
    a2 =  kuwo.loc[kuwo.album_id == id2].iloc[0]
    if a1.artist_name != a2.artist_name and not (a1.artist_name in a2.artist_name or a2.artist_name in a1.artist_name):
        if a1.artist_name not in filter_artist:
            print(id1,id2,a1.album_name, a2.album_name,a1.artist_name, a2.artist_name)
            filter_list.append([id1,id2])
filter_list.append([86631766,2176574])
filter_df = pd.DataFrame(filter_list)
filter_df.columns = ['album_id1', 'album_id2']
```

```python
filter_df
```

```python
q12 = """select r11.*
from r11
where (r11.album_id1, r11.album_id2) not in (select * from filter_df)"""
```

```python
same_album = pysqldf(q12)
same_album.to_csv("same_entity_pair/same_album",header = False,index=False, sep='\t')
same_album
```

```python
q4= """
select distinct cloud.song_name as song_name1,kuwo.song_name as song_name2,cloud.song_id as song_id1, kuwo.song_id as song_id2
from cloud,kuwo
where cloud.song_name != kuwo.song_name 
and abs(cloud.song_duration - kuwo.song_duration) < 5
and cloud.album_name = kuwo.album_name
and cloud.artist_name = kuwo.artist_name
and cloud.song_name like '%-%'"""
```

```python
r4= pysqldf(q4)
li = []
for i in range(r4.shape[0]):
    if r4.iloc[i].song_name2 in r4.iloc[i].song_name1:
        li.append([r4.iloc[i].song_id1,r4.iloc[i].song_id2])
r4 = pd.DataFrame(li)
r4.columns = ['song_id1', 'song_id2']
r4
```
