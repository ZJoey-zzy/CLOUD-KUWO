```python
cloud_id2artist_df = pd.read_csv("data/cloud_id2artist",header=None, encoding='utf-8',delimiter='\t')
kuwo_id2artist_df = pd.read_csv("data/kuwo_id2artist", header=None, encoding='utf-8',delimiter='\t')
```

```python
def df_to_dict(df):
    id2artist = {}
    artist2id = {}
    for i in range(df.shape[0]):
        id_, artist= df.iloc[i]
        id2artist[id_] = artist
        artist2id[artist] = id_
    return id2artist, artist2id
```

```python
cloud_id2artist, cloud_artist2id = df_to_dict(cloud_id2artist_df)
kuwo_id2artist, kuwo_artist2id = df_to_dict(kuwo_id2artist_df)
```

```python
def sorted_artist_name(artist_name):
    artist_li = artist_name.split('/')
    while '' in artist_li:
        artist_li.remove('')
    return "/".join(sorted([artist.strip() for artist in artist_li]))

def artists2ids_li(artist_name,artist2id):
    li = []
    artist_li = artist_name.split('/')
    for artist in artist_li:
        if artist not in artist2id:
            return 0
        else:
            li.append(artist2id[artist])
    return li

def song_duration_map(song_duration):
    m, s = [int(s) for s in song_duration.split(":")]
    return 60*m + s
```

```python
def filter_data(df,id2artist_df):
    id2artist, artist2id = df_to_dict(id2artist_df)
    df.artist_name = df.artist_name.map(sorted_artist_name)
    df.artist_id = df.artist_name.apply(artists2ids_li, artist2id = artist2id)
    df.song_duration = df.song_duration.map(song_duration_map)
    return df.loc[df['artist_id']!=0]
```

```python
cloud = filter_data(cloud,cloud_id2artist_df)
cloud.to_csv("prossed_cloud.csv",index=False, encoding = 'utf_8_sig')
```

```python
kuwo = filter_data(kuwo,kuwo_id2artist_df)
kuwo.to_csv("prossed_kuwo.csv",index=False, encoding = 'utf_8_sig')
```
